```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('users.data', sep='\t',names= column_names)
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
```

```python
df.head()
```

```python
movie_titles.head(15)
```

```python
movie_titles['year'] = movie_titles['title'].str.extract(r'\((\d{4})\)', expand=False)
movie_titles.isnull().sum()
```

```python
movie_titles[movie_titles.isnull().any(axis=1)]
```

```python
movie_titles.drop(movie_titles[movie_titles['item_id'] == 267].index,inplace = True)
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
df.drop("timestamp",axis=1,inplace =True)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.sort_values("rating_oy_sayisi",ascending=False).head(25)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
movie_user_ratings = moviemat['Titanic (1997)']
```

```python
similar_to_movie = moviemat.corrwith(movie_user_ratings)
```

```python
corr_movies = pd.DataFrame(similar_to_movie, columns=['Correlation'])
```

```python
corr_movies.isnull().sum()
```

```python
corr_movies = corr_movies.join(ratings['rating_oy_sayisi'])
```

```python
corr_movies[corr_movies['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head(10)
```
